# Frequency Modulation(FM)

```python
t=np.arange(0, 1.0, 0.001)

modulationIndex = 1.
modulatorFrequency=4
x_m=np.sin((2*np.pi*modulatorFrequency*t))


carrierFrequency=40; 
x_c=np.sin(2*np.pi*carrierFrequency*t); 


y_t = np.sin(2. * np.pi * (carrierFrequency * t + modulationIndex * np.sin(2*np.pi*modulatorFrequency*t)))


fig = plt.figure(figsize = (12, 8))
fig.suptitle('Frequency Modulation', fontsize=18)

plt.subplot(3,1,1)
plt.plot(t,x_m,'g')
plt.title("Modulator Signal")


plt.subplot(3,1,2)
plt.plot(t,x_c,'b')
plt.title("Carrier Signal")


plt.subplot(3,1,3)
plt.plot(t,y_t,'k')
plt.title("Frequency Modulated Signal")


fig.tight_layout(pad=2.0)
plt.show()
```

# Amplitude Modulation(AM)


```python
t=np.arange(0, 1.0, 0.001)

carrierAmplitude = 2
carrierFrequency=40; 
x_c=carrierAmplitude*np.sin(2*np.pi*carrierFrequency*t); 

modulationIndex = 1.0
phase = 0
modulatorFrequency=4
x_m=carrierAmplitude*modulationIndex*np.sin((2*np.pi*modulatorFrequency*t))

y_t = (1+modulationIndex*np.sin((2*np.pi*modulatorFrequency*t) + phase))*carrierAmplitude*np.sin(2*np.pi*carrierFrequency*t)


fig = plt.figure(figsize = (12, 8))
fig.suptitle('Amplitude Modulation', fontsize=18)

plt.subplot(3,1,1)
plt.plot(t,x_m,'g')
plt.title("Modulator Signal")


plt.subplot(3,1,2)
plt.plot(t,x_c,'b')
plt.title("Carrier Signal")


plt.subplot(3,1,3)
plt.plot(t,y_t,'k')
plt.title("Amplitude Modulated Signal")


fig.tight_layout(pad=2.0)
plt.show()
```

# Phase Modulation(PM)

```python
t=np.arange(0, 1.0, 0.001)
phase = 0

carrierAmplitude = 2
carrierFrequency=20 
x_c=carrierAmplitude*np.sin(2*np.pi*carrierFrequency*t); 

amplitudeSensitivity = 1.0
modulatorAmplitude = 1
modulatorFrequency=30
x_m= modulatorAmplitude*np.sin((2*np.pi*modulatorFrequency*t))

y_t = carrierAmplitude*np.sin((2*np.pi*carrierFrequency*t) + (amplitudeSensitivity*x_m + phase))


fig = plt.figure(figsize = (12, 8))
fig.suptitle('Phase Modulation', fontsize=18)

plt.subplot(3,1,1)
plt.plot(t,x_m,'g')
plt.title("Modulator Signal")


plt.subplot(3,1,2)
plt.plot(t,x_c,'b')
plt.title("Carrier Signal")


plt.subplot(3,1,3)
plt.plot(t,y_t,'k')
plt.title("Phase Modulated Signal")


fig.tight_layout(pad=2.0)
plt.show()
```
