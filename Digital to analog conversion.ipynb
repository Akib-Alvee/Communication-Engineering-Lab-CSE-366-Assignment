$\textbf{A program to simulate the digital-analog conversion techniques (ASK,FSK, PSK)}$

**Digital-to-analog conversion** is the process of changing one of the characteristics of an analog signal based on the information in digital data.There are three mechanisms for modulating digital data into an analog signal: **Amplitude Shift Keying (ASK)**, **Frequency Shift Keying (FSK)** and **Phase Shift Keying (PSK)**.

```python
# Import necessary library
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
sns.set()
```

# Amplitude Shift Keying (ASK)

When the amplitude of the signal is 1, the amplitude of the carrier frequency is
held; when the amplitude of the signal is 0, the amplitude of the carrier frequency
is zero.

```python
inputStream = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0, 1]

timeInterval = 10e-1
T = np.linspace(0,timeInterval,100)
signal = np.array([])
N = len(inputStream)

for i in range (N) :
    temp = inputStream[i]*np.ones(len(T))
    signal = np.append(signal,temp)
t = np.linspace(0,N*timeInterval,len(signal))

fig = plt.figure(figsize = (12, 8))
fig.suptitle('Amplitude Shift Keying', fontsize=18)

plt.subplot(3,1,1)
plt.plot(t,signal,'g')
plt.title("Message Signal")
plt.ylabel('X(t)')

plt.subplot(3,1,2)
y1=np.cos(2*np.pi*10*t)
plt.plot(t,y1,'b')
plt.title("Carrier Signal")
plt.ylabel('C(t)')

plt.subplot(3,1,3)
mult=(signal*y1)
plt.plot(t,mult,'k')
plt.title("ASK Modulated Signal")
plt.ylabel('Y(t)')

fig.tight_layout(pad=2.0)
plt.show()
```

# Frequency Shift Keying (FSK)

We have selected two carrier frequencies, frequency1 and frequency2. We use the first carrier if
the data element is 0; we use the second if the data element is 1.

```python
inputStream = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0, 1]
frequency1 = 2
frequency2 = 10

timeInterval = 10e-1
T = np.linspace(0,timeInterval,100)
signal = np.array([])
N = len(inputStream)

for i in range (N) :
    temp = inputStream[i]*np.ones(len(T))
    signal = np.append(signal,temp)
t = np.linspace(0,N*timeInterval,len(signal))

fig = plt.figure(figsize = (12, 8))
fig.suptitle('Frequency Shift Keying', fontsize=18)

plt.subplot(4,1,1)
plt.plot(t,signal,'g')
plt.title("Message Signal")
plt.ylabel('X(t)')

plt.subplot(4,1,2)
y1=np.cos(2*np.pi*frequency1*t)
plt.plot(t,y1,'b')
plt.title("Carrier Signal: 01")
plt.ylabel('C1(t)')

plt.subplot(4,1,3)
y2=np.cos(2*np.pi*frequency2*t)
plt.plot(t,y2,'c')
plt.title("Carrier Signal: 02")
plt.ylabel('C2(t)')

fskModulatedSignal = np.array([])
for i in range(N):
  if inputStream[i] == 1:
    y1=np.cos(2*np.pi*frequency1*T)
    fskModulatedSignal = np.append(fskModulatedSignal,y1)
  else:
    y2=np.cos(2*np.pi*frequency2*T)
    fskModulatedSignal = np.append(fskModulatedSignal, y2)

plt.subplot(4,1,4)
plt.plot(t,fskModulatedSignal,'k')
plt.title("FSK Modulated Signal")
plt.ylabel('Y(t)')

fig.tight_layout(pad=1.0)
plt.show()
```

# Phase Shift Keying (PSK)

we have only two signal elements, one with a phase of 0°, and the other with a phase of 180°.

```python
inputStream = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0, 1]

timeInterval = 10e-1
T = np.linspace(0,timeInterval,100)
signal = np.array([])
N = len(inputStream)

for i in range (N) :
    temp = inputStream[i]*np.ones(len(T))
    signal = np.append(signal,temp)
t = np.linspace(0,N*timeInterval,len(signal))

fig = plt.figure(figsize = (12, 8))
fig.suptitle('Phase Shift Keying', fontsize=18)

plt.subplot(3,1,1)
plt.plot(t,signal,'g')
plt.title("Message Signal")
plt.ylabel('X(t)')

plt.subplot(3,1,2)
y1=np.sin(2*np.pi*2*t)
plt.plot(t,y1,'b')
plt.title("Carrier Signal")
plt.ylabel('C(t)')

pskModulatedSignal = np.array([])
for i in range(N):
  if inputStream[i] == 1:
    temp=np.sin(2*np.pi*2*T)
    pskModulatedSignal = np.append(pskModulatedSignal,temp)
  else:
    temp=np.sin(2*np.pi*2*T + np.pi)
    pskModulatedSignal = np.append(pskModulatedSignal, temp)

plt.subplot(3,1,3)
mult=(signal*y1)
plt.plot(t,pskModulatedSignal,'k')
plt.title("PSK Modulated Signal")
plt.ylabel('Y(t)')

fig.tight_layout(pad=2.0)
plt.show()
```
