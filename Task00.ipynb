$$ \textbf{Signal-01: Signals and Plotting the signal.} $$


In  **Data Communication**, information converted into an electrical form suitable for transmission is called a signal.In data communications, periodic analog signals and nonperiodic digital signals are are commonly used. A periodic signal completes a pattern within a measurable time frame, called a
period, and repeats that pattern over subsequent identical periods.The simplest way to show signals is by plotting them on a pair of perpendicular axes. The vertical axis represents the value or strength of a signal. The horizontal axis represents time.

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns 
sns.set()
```

A sine wave or sinusoidal wave is the most natural representation of how many things in nature change state. A sine wave shows how the amplitude of a variable changes with time.
$$x(t) = A\sin(2\pi f t + phase)$$

```python
t = np.linspace(0, 2.0, 150)
def create_sinusoidal(A, f, t, phase = 0):
  x = A * np.sin(2.0 * np.pi * f * t + phase)
  return x
```

```python
x = create_sinusoidal(1,1,t)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
fig.suptitle('\nDifferent subplots of a sinusoidal signal', fontsize=15)

ax[0, 0].plot(t,x)
ax[0, 0].set_title('Continuous plot', fontsize = 15)

ax[0, 1].stem(t,x)
ax[0, 1].set_title('Stem plot', fontsize = 15)

ax[1,0].scatter(t,x)
ax[1,0].set_title('Scatter plot', fontsize = 15)
ax[-1,-1].axis('off')
for a in ax.flat:
    a.set(xlabel='time(t)', ylabel='x(t)')

plt.show()
```

**Composite signal** is a combination of simple sine waves with different frequencies, amplitudes, and phases.A single-frequency sine wave is not useful in data communications;we need to send a composite signal, a signal made of many simple sine waves.


```python
x1 = create_sinusoidal(1,3,t)
x2 = create_sinusoidal(2,5,t)
x3 = create_sinusoidal(3,7,t)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15, 15))
fig.suptitle('3 Sinusoidal signal and a composite signal', fontsize=15)

ax[0, 0].plot(t,x1, 'r')
ax[0, 0].set_title('x1(t)', fontsize = 15)

ax[0, 1].plot(t,x2, 'g')
ax[0, 1].set_title('x2(t)', fontsize = 15)

ax[1,0].plot(t,x3, 'b')
ax[1,0].set_title('x3(t)', fontsize = 15)

ax[1,1].plot(t, x1 + x2 + x3)
ax[1,1].set_title('y(t) = x1(t) + x2(t) + x3(t)', fontsize = 15)
plt.show()
```

To convert a continuous time signal into discrete time signal,sampling should be done.Sampling ratre or frequency should be double or more from orginal frequency.
$$A\sin(2\pi f t + phase)=x(t)  \to x(n)$$

```python
fs = 20
step = 1.0/fs
n = np.arange(0, 2.01, step)
y = create_sinusoidal(1,1,n)
plt.title('fs = 20', fontsize = 15)
plt.stem (n,y)
plt.show()
```

**Periodicity**

A continuous signal is said to be periodic if there exists such a time interval T that x(t+T)= x(t); for all t.

```python
t = np.arange(-(4*np.pi),4*np.pi,0.1)
x_t = np.sin(2*(np.pi)+t)
X_T = np.sin(t)


fig = plt.figure(figsize = (8, 6))
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.title("Periodicity of Signal", fontsize = 15)
plt.plot(t,x_t, 'b-', linewidth=4, label='x(t+T)')
plt.plot(t,X_T, 'g-', linewidth=2, label='x(t)')


plt.legend(loc='upper left')

plt.show()
```

**Shifting of a signal**

A signal x(n) may be shifted
in time by replacing the independent variable n by n k, where k is an integer. If k
is a positive integer, the time shift results in a delay of the signal by k units of time.
If k is a negative integer, the time shift results in an advance of the signal by |k| units
in time.

```python
def Shifting(y,k,n):
  ret=np.zeros(abs(k))

  if(k>=0):
    ret=np.concatenate((ret,y[0:len(n)-k]))
  else:
    ret=np.concatenate((y[abs(k):len(n)],ret))
  return ret
```

```python
fig, ax = plt.subplots(2,2, figsize=(15, 15))
fig.suptitle('\nShifting of a signal', fontsize=15)

ax[0,0].stem(n,y)
ax[0,0].set_title("Orginal Signal", fontsize = 15)

ax[0,1].remove()

k=5
ax[1, 0].stem(n,Shifting(y,k,n))
ax[1, 0].set_title('Delay', fontsize = 15)

k=-5
ax[1, 1].stem(n,Shifting(y,k,n))
ax[1, 1].set_title('Advance', fontsize = 15)


plt.show()
```

**Folding or Reflection**

Folding or Reflection of the signal about the time origin n = 0 is the time base replacement of the independent variable n by n.

```python
plt.stem(n,y)
plt.title("Orginal Signal")
plt.show()

plt.stem(-n,y)
plt.title("Folded Signal")
plt.show()
```
